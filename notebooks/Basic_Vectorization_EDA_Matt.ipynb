```python
df = train.sample(frac=0.20, random_state=42)
```

```python
X = df['comment_text']
y = df['target']
```

```python
X = X.dropna()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
import string

stemmer = nltk.stem.PorterStemmer()


stop_words = stopwords.words('english')

# For now I am just going to filter out both 'positive' and 'negative' to see how the model performs without them both.
# nltk's stopwords already filters out 'no' from the beginning so the only disproportianately large return from the null 
# values in positive_review or negative_review would be positive or negative.
# custom_stopwords = ['negative','positive'] 

# # adding the custom wordlist to the main list
# stop_words.extend(custom_stopwords)


def my_tokenizer(sentence):
    for punctuation_mark in string.punctuation:
        sentence = sentence.replace(punctuation_mark,'').lower()
    
    listofwords = sentence.split(' ')
    listofstemmed_words = []
    
    
    for word in listofwords:
        if (not word in stop_words) and (word!=''):
            stemmed_word = stemmer.stem(word)
            listofstemmed_words.append(stemmed_word)

    return listofstemmed_words
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# establish a vectorizer for the negative reviews 
count_vec = CountVectorizer(
                        min_df=10, # don't want any words that appear less than 10 times 
                        tokenizer=my_tokenizer, # use my custom tokenizer 
                        ngram_range=(1, 3), # unigrams, bi-grams, and tri-grams are all accepted 
                        max_features=500) # only return 500 results 

count_vec.fit(X) # fit to the X_train
```

```python
X_test = test_public['comment_text']
y_test = test_public['toxicity']
```

```python
X_train_transformed = count_vec.transform(X)
X_test_transformed = count_vec.transform(X_test)

print(X_train_transformed.shape)
print(X_test_transformed.shape)
```

```python
test_df = pd.DataFrame(columns=count_vec.get_feature_names_out(), 
                           data=X_test_transformed.toarray(),
                           index=X_test.index)
display(test_df)
```

```python
train_df = pd.DataFrame(columns=count_vec.get_feature_names_out(), 
                           data=X_train_transformed.toarray(),
                           index=X.index)
display(train_df)
```
