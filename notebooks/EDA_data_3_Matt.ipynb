```python
# for loop value counts 

tables = [
    data,
    data2,
    data3,
    data4,
    data5,
    data6
]


label_value_counts = []

for table in tables:
    exploded = table['labels'].explode()  
    counts = exploded.value_counts()  
    label_value_counts.append(counts)  
```

```python
label_value_counts
```

```python
from functools import reduce


combined_counts = reduce(lambda x, y: x.add(y, fill_value=0), label_value_counts)

combined_counts.sort_values(ascending=False, inplace=True)  

combined_counts
```

The TLDR of the hugging face datasets is that there are 9 different categories in the train and validation datasets.  They are as follows:

- toxic        
- selfharm   
- insult     
- profane     
- hate          
- offensive    
- threat
- sexual
- harassment 

The labels also come pre-encoded so that an algorithm can better identify the labels within the text.  The one issue which I am uncertain at this point whether it would affect our project or not is that the languages are not consistent.  From what I can tell there are several different languages used in the text columns and at this point I'm not entirely certain how to separate them for training purposes.  On the other hand this dataset is pretty strongly optimized for machine learning.  6 training sets and one validation set combined with human legible and machine legible labels means we could probably create a fairly strong toxicity detector from this group.
